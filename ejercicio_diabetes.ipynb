```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data.dtypes
```

```python
print("Número original de datapoints\n", data.shape)
```

```python
data = data.dropna() #elimna cualquier valor faltantey reemplaza las variables de los datos
```

```python
print("Número actualizado de datapoints\n", data.shape)
```

```python
import matplotlib.pyplot as plt

#ajustar el tamaño de fuente 
plt.rcParams['font.size'] = 15

#crear una figura y ajusta su tamaño
f = plt.figure(figsize=(8,4))

# crear un subplot o subtrama, al ser una sola figura es 1,1,1.

ax = f.add_subplot(1,1,1)

#grafica tus datos usando "hist". pasa el objeto ax a pandas, agrega un borde negro con grosor de 2.
data["Outcome"].hist(ax=ax, edgecolor='black', linewidth=2)

#establece los limites en el eje x
ax.set_xlim([-0.5, 1.5])

#establece la frevuencia de tick, tenemos 0 y 1 que corresponde a sí y no respectivamente.
ax.set_xticks([0,1])

#etiquetar los xticklabels
ax.set_xticklabels(["N","Y"])

#crea el titulo.
ax.set_title("Diabetes Y/N?")

#establece la etiqueta del eje x
ax.set_xlabel("Answer")

#establece la etiqueta del eje y
ax.set_ylabel("Count")

#establece los limites superoior en inferior del eje y
ax.set_ylim([0, 600])

#hace que las cosas sean bonitas, no es necesario, pero se ajusta al tamaño de la figura
f.tight_layout()
```

```python
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
data["Age"].hist(ax=ax, edgecolor='black', linewidth=2)
ax.set_title("Age range of patients")
ax.set_ylim([0,350])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
```

```python
binvalues = [20, 25, 30, 35, 40, 85]
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
data["Age"].hist(ax=ax, bins = binvalues, edgecolor='black', linewidth=2)
ax.set_title("Age range of patients")
ax.set_ylim([0,250])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
```

```python

f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
data["Age"].hist(ax=ax, bins = 5, edgecolor='black', linewidth=2)
ax.set_title("Age range of patients")
ax.set_ylim([0,500])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
```

```python
plt.rcParams['font.size'] = 15
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
ax.scatter(data["Age"],data["BloodPressure"], alpha=0.25)
ax.set_title("Blood pressure vs age range of patients")
ax.set_ylabel("Blood Pressure")
ax.set_xlabel("Age")
f.tight_layout()
```

```python
#Seleccionar las filas donde el valor de la presión arterial sea mayor a 0, pues el que sean igual a ese valor parece indicar que no se registraron

data = data[data["BloodPressure"] > 0]
print("Patients with blood pressure \n", data.shape) #rows x columns
```

```python
plt.rcParams['font.size'] = 15
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
ax.scatter(data["Age"],data["BloodPressure"], alpha=0.25)
ax.set_title("Blood pressure vs age range of patients")
ax.set_ylabel("Blood Pressure")
ax.set_xlabel("Age")
f.tight_layout()
```

```python
diabetic = data[data["Outcome"] == 1]
notdiabetic = data[data["Outcome"] == 0]

plt.rcParams['font.size'] = 15
f = plt.figure(figsize=(8,8))
ax = f.add_subplot(1,1,1)
ax.scatter(diabetic["BloodPressure"], diabetic["Age"], alpha=0.25, c='r')
ax.set_title("Diabetic: Age range of patients vs Blood Pressure")
ax.set_xlabel("Blood Pressure")
ax.set_ylabel("Age")

plt.rcParams['font.size'] = 15
f = plt.figure(figsize=(8,8))
ax = f.add_subplot(1,1,1)
ax.scatter(notdiabetic["BloodPressure"], notdiabetic["Age"], alpha=0.25, c='r')
ax.set_title("Not Diabetic: Age range of patients vs Blood Pressure")
ax.set_xlabel("Blood Pressure")
ax.set_ylabel("Age")

f.tight_layout()
```
